```python
import load_data
import pandas as pd

# return DataSet class
data = load_data.read_data_sets(one_hot=True)

# get train data and labels by batch size
BATCH_SIZE = 32
train_x, train_label = data.train.next_batch(BATCH_SIZE)

# get test data
test_x = data.test.data

# get test labels
test_labels = data.test.labels

# get sample number
n_samples = data.train.num_examples
```

```python

# Convert training data and labels into a DataFrame
train_df = pd.DataFrame(train_x)
train_df['label'] = pd.Series(train_label.argmax(axis=1))  # Assuming one-hot encoded labels

# Convert test data and labels into a DataFrame
test_df = pd.DataFrame(test_x)
test_df['label'] = pd.Series(test_labels.argmax(axis=1))  # Assuming one-hot encoded labels

# Head of DataFrames
print("Training Data Head:")
print(train_df.head())

print("\nTest Data Head:")
print(test_df.head())
```
